## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes
   For Example "Landing_Outcome"**

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
uniqueLaunchSites = pd.read_sql_query("SELECT DISTINCT launch_site FROM spacextbl", con)
uniqueLaunchSites
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
ccaLaunches = pd.read_sql_query("SELECT * FROM spacextbl WHERE launch_site LIKE 'CCA%' LIMIT 5", con)
ccaLaunches
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
totalMassCarried = pd.read_sql_query("SELECT customer, sum(PAYLOAD_MASS__KG_) FROM spacextbl WHERE customer LIKE 'nasa%' GROUP BY customer LIKE 'nasa%' ", con)
totalMassCarried
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
averageMassCarried = pd.read_sql_query("SELECT booster_version, AVG(PAYLOAD_MASS__KG_) FROM spacextbl WHERE Booster_Version LIKE 'F9 v1.1%' GROUP BY booster_version LIKE 'F9 v1.1%'", con)
averageMassCarried
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
# Ground pad?
print( pd.read_sql("SELECT DISTINCT landing_outcome FROM spacextbl", con) )
firtsSucessfullGrounpad = pd.read_sql_query("SELECT min(date) as Date, landing_outcome as 'Landing Outcome' FROM spacextbl WHERE landing_outcome = 'Success (ground pad)' ", con)
firtsSucessfullGrounpad
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
boosterQ = pd.read_sql_query("SELECT DISTINCT booster_version FROM spacextbl WHERE landing_outcome = 'Success (drone ship)' AND PAYLOAD_MASS__KG_ BETWEEN 4000 AND 6000", con)
boosterQ
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
successAndFailure = pd.read_sql_query("SELECT COUNT(*) as 'Sucess/Failure count', landing_outcome FROM spacextbl WHERE landing_outcome LIKE 'success' OR landing_outcome LIKE 'failure%' GROUP BY landing_outcome LIKE 'success'", con)
successAndFailure
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
maximumPayloadBooster = pd.read_sql_query(
    "SELECT DISTINCT booster_version FROM spacextbl WHERE PAYLOAD_MASS__KG_ = ( SELECT max(PAYLOAD_MASS__KG_) FROM spacextbl )", con)
maximumPayloadBooster
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
display(df[:2])

records = pd.read_sql_query("SELECT booster_version, substr(Date, 6,2) as Month, Landing_Outcome from spacextbl WHERE date LIKE '2015%' AND Landing_Outcome = 'Failure (drone ship)' ", con)
records
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
rank = pd.read_sql_query("SELECT count(landing_outcome) as 'Count', date, landing_outcome as 'Good/Bad landing' FROM spacextbl WHERE date BETWEEN '2010-06-04' AND '2017-03-20' GROUP BY Landing_Outcome LIKE 'Success%'", con)
rank
```
